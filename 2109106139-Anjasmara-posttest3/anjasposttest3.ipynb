```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

masukkan data

```python
df = pd.read_csv("all_bikez_curated.csv")
```

```python
df.head()
```

grafik batang

```python
# Visualisasi Jumlah Sepeda Motor Berdasarkan Merek
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Brand')
plt.title('Jumlah Sepeda Motor Berdasarkan Merek')
plt.xlabel('Merek')
plt.ylabel('Jumlah')
plt.xticks(rotation=90)
plt.show()
plt.show()
##Dari visualisasi ini, kita dapat melihat seberapa banyak sepeda 
##motor yang ada untuk setiap merek. Merek tertentu mungkin lebih dominan dalam dataset ini.
```

pie chart

```python
# Hitung jumlah setiap kategori dalam kolom "jenis_motor"
jenis_motor_counts = df['Brand'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(jenis_motor_counts, labels=jenis_motor_counts.index, autopct='%1.1f%%')
plt.title('Sebaran Jenis Motor')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

heatmap


```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Korelasi antar Fitur pada Data Spesifikasi Sepeda Motor")
plt.xlabel("Fitur")
plt.ylabel("Fitur")
plt.show()
```

box plot

```python
# Membuat box plot untuk tahun produksi
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', data=df)
plt.title('Distribusi Tahun Produksi Motor')
plt.xlabel('Tahun Produksi')
plt.ylabel('Nilai')
plt.show()
```
